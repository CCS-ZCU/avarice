```python
import pandas as pd
import nltk
```

```python
auaritia_df = pd.read_parquet("../data/auaritia.parquet")
```

```python
# look at the first few rows
auaritia_df.head()
```

```python
# filter to focus exclusively on medieval texts
auaritia_df_medieval = auaritia_df[
    (auaritia_df["not_before"].between(1050, 1250)) |
    (auaritia_df["not_after"].between(1050, 1250))]#%%
# let's do some temporal filtering
```

```python
# we have four different versions of the textual context, each with its own advantages and disadvantages
auaritia_df_medieval[["kwic_text", "kwic_tokens", "target_sentence_text", "target_sentence_tokens"]]

```

```python
# KWIC tokens is my favorite! enormly versatile!

def get_semantic_lemmata(kwic_tokens):
    return [token["lemma"] for token in kwic_tokens if token["pos"] in ["NOUN", "VERB", "ADJ"]]
```

```python
auaritia_df_medieval["kwic_tokens"].apply(get_semantic_lemmata)
```

```python
kwic_lemmata_flat = [w for c in auaritia_df_medieval["kwic_tokens"].apply(get_semantic_lemmata) for w in c]
kwic_lemmata_flat[:10]
```

```python
nltk.FreqDist(kwic_lemmata_flat).most_common(50)
```
